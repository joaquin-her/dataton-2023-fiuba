```python
#sin fecha, con horarios, relacionada con la hora de los viajes y cuantos viajes por hora
#key = 00n + NOMBRE 
import pandas as pd 
estaciones_csv = pd.read_csv("estaciones_bicis .csv")
estaciones_csv.head()
```

```python
estaciones_csv.columns
```

```python
def crearFormatID(fila):
    #La funcion devuelve la cadena concatenada de el indice con 3 cifras y el nombre
    id = str(fila[0]).zfill(3)
    return f"{str(id)} - {str(fila[2])}"
```

```python
estaciones_csv["FormatId"] = estaciones_csv.apply(crearFormatID, axis=1)
nuevo_df = estaciones_csv["FormatId"]
```

```python
nuevo_df.head()
```

```python
marzo = pd.read_csv("marzo_dataset .csv")
```

```python
marzo.head()
```

```python
marzo['fecha_origen_recorrido'] = pd.to_datetime(marzo['fecha_origen_recorrido'])
```

```python
marzo['hora_origen'] = marzo['fecha_origen_recorrido'].dt.hour
```

```python
marzo.head()
```

```python
marzo.sort_values('nombre_estacion_origen')
```

```python
origen_marzo_por_hora = marzo[['Id_recorrido','nombre_estacion_origen','hora_origen']]
origen_marzo_por_hora.head()
```

```python
origen_marzo_por_hora.sort_values('nombre_estacion_origen')
```

```python
origen_marzo_por_hora.groupby(['nombre_estacion_origen', 'hora_origen']).size().reset_index(name='viajes_por_hora_y_estacion')
```
